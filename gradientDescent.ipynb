# Gradient Dresent as a Function

```python
def gradient_desecent(derivative_func,initial ,multiplier=0.02,precision=0.001,max_iter=300):
    
    new_x=initial

   

    x_list=[new_x]
    slope_list=[derivative_func(new_x)]

    for n in range(max_iter):
        previous_x=new_x
        gradient=derivative_func(previous_x)
        new_x=previous_x-multiplier *gradient
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
#         print('new_x',new_x)
#         print('  ')
        step_size=abs(new_x-previous_x)
#         print('step',step_size)


        if(step_size<precision):
        
            break
    return new_x,x_list,slope_list        

   
#     print('Slope value at this point is:',df(new_x))  
#     print('f(x) or cost at this point is:',f(new_x))  
    #a steep slope indicates an increase in error the closer we are to the minimum slope the lesser the erro
```

```python
local_min,list_x,deriv_list =gradient_desecent(derivative_func=dg,initial=0.5,multiplier=0.02,precision=0.001)
print('Local minimum occurs at:',local_min) 
print('number of steps',len(list_x)) 
```

# Batch Gradient Descent with Sympy

```python
# setup
multiplier=0.1
max_iter =200
params = np.array([1.0,1.0])
for n in range(max_iter):
    gradient_x=diff(f(a,b),a).evalf(subs={a:params[0], b:params[1]})
    gradient_y=diff(f(a,b),b).evalf(subs={a:params[0], b:params[1]})
    gradients = np.array([gradient_x,gradient_y])
    params=params - multiplier * gradients
    
print('values in gradient array',gradients)
print('minimum occurs at x values of', params[0])
print('minimum occurs at y values of', params[1])
print('the cost is of', f(params[0],params[1]))
    
```

```python
# partial derviative function
def fpx(x,y):
    r=3**(-x**2 -y**2)
    return 2*x*log(3)*r/(r+1)**2
```

```python
def fpy(x,y):
    r=3**(-x**2 -y**2)
    return 2*y*log(3)*r/(r+1)**2
```

# Example 5 - Working with Data and Real cost function
## Mean squared error: a cost function for regression problems

### $$RSS=\sum_{i=1}^{n}\big(y^{(i)} -h_\theta x^{(i)} \big)^2$$
### $$MSE=\frac{1}{n}\sum_{i=1}^{n}\big(y^{i} -h_\theta x^{{i}} \big)^2$$

### $$MSE=\frac{1}{n}\sum_{i=1}^{n}\big(y -\hat{y} \big)^2$$
## $$\hat{y}=\theta_0+\theta_1x$$

```python
#making data
x_5=np.array([[0.1,1.2,2.4,3.2,4.1,5.7,6.5]]).transpose()
y_5=np.array([[1.7,2.4,3.5,3.0,6.1,9.4,8.2]]).transpose()
print(x_5.shape) #flatten array no row or column
print(y_5.shape)
```

```python
# quick linear regression
# regression fit does not take a 1D array
regr=LinearRegression()
regr.fit(x_5,y_5)
print('Theta 0:', regr.intercept_[0])
print('Theta 1:', regr.coef_[0][0])
```

```python
plt.scatter(x_5,y_5,s=50) # actual points
plt.plot(x_5,regr.predict(x_5),color='orange',linewidth=3) #fitted points
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
# y_hat = theta0 +theta1 *x
y_hat= 0.8475351486029536+1.2227264637835915*x_5 # fitted
```

```python
print(y_hat)
```

```python
def MSE(y,y_hat):
    mse_calc =(1/y.size)*sum((y-y_hat)**2)
    return mse_calc
    
```

```python


print('Mse with function',MSE(y_5,y_hat))
print('Mse with sklearn', mean_squared_error(y_5,y_hat))
print('Mse ', mean_squared_error(y_5,regr.predict(x_5)))
```

## partial derviatives wrt $\theta_0$ and $\theta_1$
## $$\frac{\partial MSE}{\partial \theta_0}=-\frac{2}{n}\sum_{i=i}^{n}\big( y^{(i)}-{\theta_0}-{\theta_1 x^{(i)}}\big)$$

## $$\frac{\partial MSE}{\partial \theta_1}=-\frac{2}{n}\sum_{i=i}^{n}\big( y^{(i)}-{\theta_0}-{\theta_1 x^{(i)}}\big)\big(x^{(i)}\big)$$

```python
def grad(x,y,thetas):
    n=y.size
    theta0_slope=(-2/n)*sum(y-thetas[0] - thetas[1] *x)
    theta1_slope=(-2/n)*sum(y-thetas[0] - thetas[1] *x)*x
    return np.array(theta0_slope,theta1_slope)
grad(x_5,y_5,thetas)
```

```python
multiplier=0.01
thetas=np.array([2.9,2.9])
plot_values=thetas.reshape(1,2)
mse_vals = MSE(y_5,thetas[0]+thetas[1]*x_5)
for i in range(1000):
    thetas = thetas -multiplier *grad(x_5,y_5,thetas)
    plot_values=np.concatenate(plot_values,thetas.reshape(1,2),axis=0)
    mse_vals=np.append(arr=mse_vals, values=MSE(y_5,thetas[0]+thetas[1]*x_5),axis=0)
#results
print('min occurs at theta 0 is at:', thetas[0])
print('min occurs at theta 1 is at:',thetas[1])
print('MSE is:',MSE(y_5,thetas[0]+thetas[i]*x_5))
```
